## 作業目標

根據範例：

1. 取出今天所有的新聞
2. 取出現在時間兩小時內的新聞
3. 根據範例，取出三天前下午三點到五點的新聞

## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup

# 打開瀏覽器
browser = webdriver.Chrome(executable_path='./chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")


selectYear = Select(browser.find_element_by_id("selY"))
selectYear.select_by_value('2020')

selectMonth = Select(browser.find_element_by_id("selM"))
selectMonth.select_by_value('3')

selectDay = Select(browser.find_element_by_id("selD"))
selectDay.select_by_value('30')

browser.find_element_by_id('button').click()
```

```python
# 每隔兩秒鐘自動往下滑
import time
for i in range(100):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")
```

```python
# 取得資料，丟到 BeautifulSoup 解析

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

browser.quit();
```

```python
import re
from datetime import date

today = date.today()

d1 = today.strftime("%Y/%m/%d")
for d in soup.find(class_="part_list_2").find_all('h3'):
    if d.find(text=re.compile(d1)) != None:
        print(d.find(text=re.compile(d1)), d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
from datetime import datetime, timedelta
hours_from_now = datetime.now() - timedelta(hours = 2)
time = hours_from_now.strftime('%Y/%m/%d %H:%M')
print("Time compared:", time)

for d in soup.find(class_="part_list_2").find_all('h3'):
    Date = datetime.strptime(d.find(class_="date").text,'%Y/%m/%d %H:%M')
    if hours_from_now.time() <= Date.time() and d.find(text=re.compile(d1)) != None:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
# 打開瀏覽器
browser = webdriver.Chrome(executable_path='./chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

days_from_now = datetime.now() - timedelta(days = 3)

year = days_from_now.strftime('%Y')
month = str(int(days_from_now.strftime('%m')))
day = days_from_now.strftime('%d')


selectYear = Select(browser.find_element_by_id("selY"))
selectYear.select_by_value(year)

selectMonth = Select(browser.find_element_by_id("selM"))
selectMonth.select_by_value(month)

selectDay = Select(browser.find_element_by_id("selD"))
selectDay.select_by_value(day)

browser.find_element_by_id('button').click()
```

```python
# 每隔兩秒鐘自動往下滑
import time
for i in range(150):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")
```

```python
# 取得資料，丟到 BeautifulSoup 解析

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

browser.quit();
```

```python
import re
from datetime import datetime, date, time

three = datetime(int(year), int(month), int(day),15, 0)
five = datetime(int(year), int(month), int(day),17, 0)
for d in soup.find(class_="part_list_2").find_all('h3'):
    Date = datetime.strptime(d.find(class_="date").text,'%Y/%m/%d %H:%M')
    if three < Date and Date < five:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```
